### Manejo XML en Python

```python
from xml.dom import minidom
```

```python
xmldoc = minidom.parse('./data/05_XML.xml')
```

```python
articulos = xmldoc.getElementsByTagName('articulo')
```

```python
for art in articulos:
    print(art.attributes['nombre'].value)
```
