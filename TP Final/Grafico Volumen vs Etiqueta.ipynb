```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel('C:/Users/ferna/Downloads/YPF Cotizaciones.xlsx', sheet_name= 'ACIERTOS')
```

```python
fechas = data['Date']
valores_acciones = data['Volume']

resultados_modelo = data['ACIERTOS']

fechas = pd.to_datetime(fechas)

dias = range(1, len(fechas) + 1)

fig, ax = plt.subplots()

ax.plot(dias, valores_acciones, label='Valores de Acciones')

for dia, resultado in zip(dias, resultados_modelo):
    color = 'red' if resultado == 0 else 'lightgreen'
    ax.axvspan(dia - 0.5, dia + 0.5, facecolor=color, alpha=0.3)

ax.set_xlabel('Fecha')
ax.set_ylabel('Valor de Acciones')
ax.set_title('YPF')

ax.legend()
plt.show()
```

```python
fechas = data['Date']
valores_acciones = data['Volume']
resultados_modelo = data['ACIERTOS']

fechas = pd.to_datetime(fechas)

dias = range(1, len(fechas) + 1)

fig, ax = plt.subplots()

ax.plot(dias, valores_acciones, label='Valores de Acciones', color='orange')

for dia, resultado in zip(dias, resultados_modelo):
    if resultado == 1:
        ax.axvspan(dia - 0.5, dia + 0.5, facecolor='lightgray', alpha=0.3)

ax.set_xlabel('Fecha')
ax.set_ylabel('Volumen')
ax.set_title('YPF')

n = len(fechas) // 10
plt.xticks(dias[::n], [fecha.strftime('%d-%m-%y') for fecha in fechas[::n]], rotation=45)

plt.show()
```
